```python
def timeline_chart(df):
    data = [
        dict(type='scatter',
             y=df.loc[df['continent'] == continent]['CO2_emissions'],
             x=df.loc[df['continent'] == continent]['year'],
             text=df.loc[df['continent'] == continent]['country_name'],
             mode='markers',
             name=continent) 
        for continent in df['continent'].unique()
    ]
```
