```python
def timeline_chart(df, cols=None, top=5, years=[], months=[]):

    df_filter = df
    if years != []: df_filter = df_filter[df_filter.index.isin(years, level=0)]
    if months != []: df_filter = df_filter[df_filter.index.isin(months, level=1)]
    
    top_types = df_filter.groupby(['Primary Type'])['Count'].sum().nlargest(top).index
    top_df = df_filter[df_filter['Primary Type'].isin(top_types)]
    pivot_df = top_df.groupby(['yyyy-mm','Primary Type'])['Count'] \
                    .sum() \
                    .to_frame() \
                    .reset_index() \
                    .pivot(index='yyyy-mm', columns='Primary Type', values='Count')
    
    return pivot_df    
```

```python
test_df = timeline_chart(df, top=3, years=[2014,2016])
test_df.head()

test_data = [
    dict(
        type='scatter',
        x=test_df.index,
        y=test_df[col],
        name=col
    )
    for col in test_df.columns
]

test_layout = go.Layout(
    xaxis=dict(
        tickformatstops=[
            dict(dtickrange=[None, "M12"], value='%Y-%b'),
            dict(dtickrange=["M12", None], value='%Y')
        ]
    )
)

test_fig = go.Figure(data=test_data, layout=test_layout)
test_fig.show()
```
